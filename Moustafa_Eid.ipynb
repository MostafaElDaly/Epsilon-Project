```python
## reading from the file
df = pd.read_csv('Google-Playstore.csv')
```

```python
# Get Release Year & Month & Day
df['Released'] = pd.to_datetime(df['Released'] )
df['year'] = df['Released'].dt.year.astype("Int64")
df['month'] = df['Released'].dt.month.astype("Int64")
df['day'] = df['Released'].dt.day.astype("Int64")
```

```python
# Number of category
df['Category'].value_counts().nunique()
```

```python
#Count the number of apps in each 'Category'. 
num_category = df['Category'].value_counts()
num_category
```

```python
num_category.plot(kind='barh')
```

```python
sns.histplot(data = df, x = 'Category', bins = 10, kde = True)
```

```python
#check Category with the mean of  Rate

d= df.groupby('Category')['Rating'].mean().sort_values(ascending = False)
d
```

```python
sns.boxplot(data = df, x = 'Rating', y = 'Category' )
```

```python
x = df.groupby('App Name')['Rating'].max().sort_values(ascending = False)
x
```

```python
sns.boxplot(data = df, x = 'Rating', y = 'App Name' )
```

```python
# APPS with Rate 5 (Best Free Apps)
d = df[df['Rating'] == 5]
top_free = d[['App Name' , 'Free']]
top_free
```

```python
# Apps with high Price
high_price = df[df['Price'] > 100]
high_price[['Category', 'App Name', 'Price']]
```

```python
sns.boxplot(x='Price', y='Category', data=df[df['Price'] > 100])
```
